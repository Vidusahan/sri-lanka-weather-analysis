# Descriptive Analysis of Sri Lanka Weather Dataset

## Introduction
Descriptive analysis is the process of summarizing and exploring data using
statistical measures, charts, and graphs. In this project, we analyze
the Sri Lanka Weather Dataset, which contains daily weather observations
from 30 cities across the country between **2010 and 2023**.

The goal of this analysis is to:
- Understand temperature, rainfall, and wind distributions.
- Identify seasonal and geographical patterns.
- Explore relationships between key weather indicators.

This section provides a first look into the dataset with summary statistics
and visualizations.


```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load dataset
DATA_PATH = '../data/raw/sri_lanka_weather_data.csv'

df = pd.read_csv(DATA_PATH)
```

```python
#Display first five rows
df.head()
```

```python
# Shape of dataset
print("Dataset Shape:", df.shape)

# Data types and missing values
print("\nDataset Info:")
print(df.info())

# Descriptive statistics for numerical columns
df.describe().T
```

##Data Cleaning & Preprocessing

```python
# Convert date columns to datetime
df['time'] = pd.to_datetime(df['time'], errors='coerce')

# Convert sunrise and sunset to datetime (keep only time part)
df['sunrise'] = pd.to_datetime(df['sunrise'], errors='coerce').dt.time
df['sunset'] = pd.to_datetime(df['sunset'], errors='coerce').dt.time

# Check for duplicates
duplicates = df.duplicated().sum()

# Check missing values
missing_summary = df.isnull().sum()

# Display results
print("Number of duplicates:", duplicates)
print("\nMissing values per column:\n", missing_summary[missing_summary > 0])
```

```python
for col in df.columns:
  print(f"{col}: {df[col].nunique()} unique values")
```

```python
# Drop less useful columns for EDA (keep a clean version for analysis)
df_eda = df.drop(columns=['sunrise', 'sunset', 'country','snowfall_sum','precipitation_sum'])
```

```python
df_eda['city'] = df_eda['city'].astype('category')
df_eda['weathercode'] = df_eda['weathercode'].astype('category')
```

## Variable Renaming

To make analysis easier, we renamed the datasetâ€™s columns
into shorter and more readable forms.
For example:
- `temperature 2m mean` â†’ `temp_mean`
- `shortwave radiation sum` â†’ `radiation`
- `et0 fao evapotranspiration` â†’ `evapotranspiration`

This improves code readability and prevents errors during analysis.


```python
df_eda = df_eda.rename(columns={
    "temperature_2m_max": "temp_max",
    "temperature_2m_min": "temp_min",
    "temperature_2m_mean": "temp_mean",
    "apparent_temperature_max": "app_temp_max",
    "apparent_temperature_min": "app_temp_min",
    "apparent_temperature_mean": "app_temp_mean",
    "shortwave_radiation_sum": "radiation",
    "rain_sum": "rain",
    "precipitation_hours": "precip_hours",
    "windspeed_10m_max": "wind_max",
    "windgusts_10m_max": "wind_gusts",
    "winddirection_10m_dominant": "wind_dir",
    "et0_fao_evapotranspiration": "evapotranspiration",
    "latitude": "lat",
    "longitude": "lon",
    "elevation": "elev",
    "weathercode": "weather_code",
    "city": "city",
    "time": "date"
})
```

##Univariate Analysis

**Temperature Distribution**



```python
plt.figure(figsize=(8,5))
sns.histplot(df_eda['temp_mean'], bins=40, kde=True, color='skyblue')
plt.title("Distribution of Mean Daily Temperature (Â°C)")
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Frequency")
plt.show()
```

- The mean daily temperature follows a near-normal distribution.  
- Most values are between **25Â°C and 30Â°C**, peaking around 27Â°C.  
- Very few days recorded below 15Â°C (hill-country) or above 35Â°C (coastal/lowland).  


**Temperature Boxplot**

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df_eda['temp_mean'], color="orange")
plt.title("Boxplot of Mean Daily Temperature")
plt.xlabel("Temperature (Â°C)")
plt.show()
```

- The median temperature is around **27Â°C**.  
- Outliers exist at both ends, reflecting cooler hill stations (e.g., Nuwara Eliya)
  and hotter lowland areas (e.g., Batticaloa).  
- The spread is relatively small, indicating stable temperatures across the dataset.  


**Rainfall Distribution**

```python
plt.figure(figsize=(8,5))
sns.histplot(df_eda['rain'], bins=40, color="blue")
plt.yscale("log")
plt.title("Distribution of Daily Rainfall (mm)")
plt.xlabel("Rainfall (mm)")
plt.ylabel("Frequency (log scale)")
plt.show()
```

- The rainfall distribution is highly skewed.  
- Most days record **0â€“10 mm rainfall**.  
- Extreme values (above 100 mm/day) occur during **monsoon periods**,
  showing heavy rain events.  


**Windspeed Distribution**

```python
plt.figure(figsize=(8,5))
sns.histplot(df_eda['wind_max'], bins=40, kde=True, color="purple")
plt.title("Distribution of Maximum Windspeed (km/h)")
plt.xlabel("Windspeed (km/h)")
plt.ylabel("Frequency")
plt.show()
```

- Windspeed is mostly between **10â€“30 km/h**, typical of Sri Lankaâ€™s climate.  
- A secondary peak occurs around 40â€“50 km/h, possibly due to coastal winds.  
- Rare extreme gusts above 60 km/h likely represent storm/cyclone events.  


### Weather Code

The dataset includes a categorical variable `weather_code`
that encodes daily weather conditions.  
Key values include:

- **0â€“3**: Clear or cloudy skies  
- **45, 48**: Fog  
- **51â€“67**: Drizzle and rain  
- **80â€“82**: Rain showers  
- **95â€“99**: Thunderstorms  

Snow-related codes (71â€“77, 85â€“86) are not relevant to Sri Lankaâ€™s
tropical climate.  

This helps interpret the frequency plot of weather codes
and connect them to real-world conditions.


**Weather Code Frequency**

```python
plt.figure(figsize=(10,5))
sns.countplot(x="weather_code", data=df_eda,
              order=df_eda['weather_code'].value_counts().index,
              palette="viridis")
plt.title("Frequency of Weather Codes")
plt.xlabel("Weather Code")
plt.ylabel("Count")
plt.show()
```

- The most common weather codes represent **clear skies and light rain**.  
- Cloudy and moderate rain conditions also appear frequently.  
- Severe weather codes (storms, heavy rain) are rare but important for disaster
  preparedness.  


##Bivariate Analysis

**Correlation Heatmap**

```python
plt.figure(figsize=(12,10))

# Select only numeric columns
corr = df_eda.select_dtypes(include=['number']).corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, cbar=True)
plt.title("Correlation Heatmap of Numerical Variables")
plt.show()
```

- **Temperature variables** are strongly positively correlated with each other.  
- **Radiation** shows a strong positive correlation with temperature.  
- **Rainfall** has weak or negative correlation with temperature (rain often occurs at moderate temps).  
- **Windspeed** and **evapotranspiration** have mild positive correlations with temperature.  


**Temperature vs Rainfall**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='temp_mean', y='rain', data=df_eda, alpha=0.3)
plt.title("Temperature vs Rainfall")
plt.xlabel("Mean Temperature (Â°C)")
plt.ylabel("Rainfall (mm)")
plt.show()
```

- Most days with **low rainfall** occur across all temperature ranges.  
- Heavy rainfall (>100 mm) tends to occur at **moderate temperatures (25â€“28Â°C)**.  
- Very high temperatures (>33Â°C) usually correspond to **dry days**.  


**Average Temperature by City**

```python
plt.figure(figsize=(12,6))
city_temp = df_eda.groupby('city')['temp_mean'].mean().sort_values()
sns.barplot(x=city_temp.index, y=city_temp.values, palette="coolwarm")
plt.xticks(rotation=90)
plt.title("Average Temperature by City")
plt.ylabel("Mean Temperature (Â°C)")
plt.show()
```

- Hill-country cities (e.g., **Hatton**) record the **lowest average temperatures**.  
- Coastal/lowland cities (e.g., **Jaffna, Trincomalee**) show the **highest averages**.  
- Overall, geography (elevation and proximity to coast) strongly influences city temperature.  


**Average Rainfall by City**

```python
plt.figure(figsize=(12,6))
city_rain = df_eda.groupby('city')['rain'].mean().sort_values()
sns.barplot(x=city_rain.index, y=city_rain.values, palette="crest")
plt.xticks(rotation=90)
plt.title("Average Rainfall by City")
plt.ylabel("Average Rainfall (mm)")
plt.show()
```

- Western and Southern cities (e.g., **Colombo, Rathnapura**) record **higher rainfall** due to monsoons.  
- Eastern and Northern cities (e.g., **Trincomalee, Jaffna**) have **lower averages**, reflecting their drier climate.  
- Rainfall distribution clearly aligns with **Sri Lankaâ€™s monsoon patterns**.  


##Temporal Analysis

Extract Date Parts (Year & Month)

```python
# Ensure date column is in datetime format
df_eda['date'] = pd.to_datetime(df_eda['date'])

# Create year and month columns
df_eda['year'] = df_eda['date'].dt.year
df_eda['month'] = df_eda['date'].dt.month
```

**Monthly Average Temperature Trend**

```python
monthly_temp = df_eda.groupby('month')['temp_mean'].mean()

plt.figure(figsize=(8,5))
monthly_temp.plot(marker='o', color='tomato')
plt.title("Average Monthly Temperature (2010â€“2023)")
plt.xlabel("Month")
plt.ylabel("Temperature (Â°C)")
plt.xticks(range(1,13))
plt.grid(True)
plt.show()
```

- Temperature is relatively stable across months (~26â€“29Â°C).  
- Slightly higher temperatures occur in **April** and **August**.  
- Slight dips appear around **Decemberâ€“January**, aligning with seasonal cooling.  


**Monthly Average Rainfall Trend**

```python
monthly_rain = df_eda.groupby('month')['rain'].mean()

plt.figure(figsize=(8,5))
monthly_rain.plot(marker='o', color='blue')
plt.title("Average Monthly Rainfall (2010â€“2023)")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.xticks(range(1,13))
plt.grid(True)
plt.show()
```

- Rainfall shows clear **monsoon patterns**.  
- Peaks occur in **Mayâ€“June** (Southwest Monsoon) and **Octoberâ€“November** (Northeast Monsoon).  
- Drier months: **Januaryâ€“March** and **Julyâ€“August**.  


**Yearly Temperature Trend**

```python
yearly_temp = df_eda.groupby('year')['temp_mean'].mean()

plt.figure(figsize=(10,5))
yearly_temp.plot(marker='o', color='darkorange')
plt.title("Yearly Average Temperature (2010â€“2023)")
plt.xlabel("Year")
plt.ylabel("Temperature (Â°C)")
plt.grid(True)
plt.show()
```

- Yearly mean temperature is mostly stable around **27Â°C**.  
- A slight **warming trend** is visible in later years (2016â€“2023).  


**Yearly Rainfall Trend**

```python
yearly_rain = df_eda.groupby('year')['rain'].mean()

plt.figure(figsize=(10,5))
yearly_rain.plot(marker='o', color='navy')
plt.title("Yearly Average Rainfall (2010â€“2023)")
plt.xlabel("Year")
plt.ylabel("Rainfall (mm)")
plt.grid(True)
plt.show()
```

- Yearly rainfall fluctuates significantly, reflecting variable monsoon strength.  
- Some years show **above-average rainfall** (possible flood years), while others are **drier** (possible drought years).  
- No strong long-term increase/decrease is visible, but variability is high.  


##Geographical Analysis

**Elevation vs Temperature**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='elev', y='temp_mean', data=df_eda, alpha=0.3, color='teal')
plt.title("Elevation vs Mean Temperature")
plt.xlabel("Elevation (m)")
plt.ylabel("Mean Temperature (Â°C)")
plt.show()
```

- Clear **negative relationship**: higher elevations have cooler temperatures.  
- Lowland areas (~0â€“100 m) record ~28â€“30Â°C, while hill-country (>1500 m, e.g., Nuwara Eliya) averages ~15â€“18Â°C.  
- Geography strongly explains temperature variation in Sri Lanka.  


**Elevation vs Rainfall**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='elev', y='rain', data=df_eda, alpha=0.3, color='darkblue')
plt.title("Elevation vs Rainfall")
plt.xlabel("Elevation (m)")
plt.ylabel("Rainfall (mm)")
plt.show()
```

- Rainfall shows **high variability** at different elevations.  
- Mid-elevation regions (e.g., central highlands) receive significant rainfall due to **orographic effect**.  
- Coastal lowlands show more stable but generally lower rainfall values.  


**Latitude vs Temperature**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='lat', y='temp_mean', data=df_eda, alpha=0.3, color='coral')
plt.title("Latitude vs Temperature")
plt.xlabel("Latitude (Â°)")
plt.ylabel("Mean Temperature (Â°C)")
plt.show()
```

- Slight **decline in temperature** with increasing latitude (moving northwards).  
- Southern coastal cities are slightly warmer, while northern areas are marginally cooler.  
- However, latitude effect is weaker than **elevation effect**.  

**Longitude vs Rainfall**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='lon', y='rain', data=df_eda, alpha=0.3, color='forestgreen')
plt.title("Longitude vs Rainfall")
plt.xlabel("Longitude (Â°)")
plt.ylabel("Rainfall (mm)")
plt.show()
```

- Western longitudes (Colombo, Galle) receive **higher rainfall** (Southwest monsoon).  
- Eastern regions (Batticaloa, Trincomalee) are comparatively **drier**.  
- This reflects **regional monsoon differences** across the island.
